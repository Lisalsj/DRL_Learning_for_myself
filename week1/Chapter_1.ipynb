注：下面的代码每次运行要从这里开始

```python
import gym
env = gym.make('MountainCar-v0')
print('观测空间 = {}'. format(env.observation_space))
print('动作空间 = {}'. format(env.action_space))
print('观测范围 = {} ~ {}'. format(env.observation_space.low,
env.observation_space.high))
print('动作数 = {}'. format(env.action_space.n))
```

```python
class SimpleAgent:
    def __init__(self, env):
        # 初始化，接收环境参数
        self.env = env

    def decide(self, observation):
        """
        根据观察值决策动作。
        :param observation: 包括位置 (position) 和速度 (velocity)
        :return: 动作 (action)
        """
        position, velocity = observation
        # 定义速度的下界和上界
        lb = min(-0.09 * (position + 0.25) ** 2 + 0.03,
                 0.3 * (position + 0.9) ** 4 - 0.008)
        ub = -0.07 * (position + 0.38) ** 2 + 0.07

        # 根据速度是否在范围内选择动作
        if lb < velocity < ub:
            action = 2  # 推动方向
        else:
            action = 0  # 静止或另一个方向
        return action

    def learn(self, *args):
        """
        学习方法，占位函数。不包含学习的方法，只有决策，所以不是RL的Agent
        """
        pass

# 使用环境初始化智能体
agent = SimpleAgent(env)
```

```python
def play(env, agent, render=False, train=False):
    """
    运行一回合游戏。
    :param env: 游戏环境
    :param agent: 智能体
    :param render: 是否渲染图形界面，默认为 False
    :param train: 是否训练智能体，默认为 False
    :return: 本回合的总奖励
    """
    # 初始化回合总奖励
    episode_reward = 0.0

    # 重置环境，获取初始观察值
    observation = env.reset()

    while True:
        # 渲染游戏界面
        if render:
            env.render()

        # 智能体决策动作
        action = agent.decide(observation)

        # 执行动作，获取下一状态、奖励以及是否结束
        next_observation, reward, done, _ = env.step(action)

        # 累计奖励
        episode_reward += reward

        # 如果训练模式开启，智能体进行学习
        if train:
            agent.learn(observation, action, reward, done)

        # 如果回合结束，退出循环
        if done:
            break

        # 更新当前观察值
        observation = next_observation

    # 返回本回合的总奖励
    return episode_reward
```

```python
env.seed(3) # 设置随机种子，让结果可复现
episode_reward = play(env, agent, render=True)
print('回合奖励 = {}'. format(episode_reward))
env.close() # 关闭图形界面
```

```python
import numpy as np
episode_rewards = [play(env, agent) for _ in range(100)]
print('平均回合奖励 = {}'. format(np.mean(episode_rewards)))
```
